```python
from sklearn.naive_bayes import BernoulliNB,MultinomialNB
```

```python
df = pd.read_csv("sentiment_analysis.csv")
```

```python
df.head()
```

```python
df.isna()
```

```python
vector1 = CountVectorizer(binary = True)
```

```python
vector2 =  CountVectorizer(binary = False)
```

```python
text = df['Sentence']
label = df['Sentiment']
```

```python
text
```

```python
text.shape
```

```python
label
```

```python
import nltk
from nltk.corpus import stopwords
```

```python
# Download stop words list
nltk.download('stopwords')
nltk.download('punkt')  # Download the tokenizer model
stop_words = set(stopwords.words('english'))
```

```python
stop_words
```

```python

def preprocess_text(Sentence):
    Sentence = Sentence.lower()
    tokens = nltk.word_tokenize(Sentence)
    filtered_tokens = [word for word in tokens if word not in stop_words]
    return ' '.join(filtered_tokens)

```

```python
# Apply preprocessing to all texts
text = text.apply(preprocess_text)
```

```python
text
```

```python
X1 = vector1.fit_transform(text)
```

```python
X1
```

```python
X1.shape
```

```python
X2 = vector2.fit_transform(text)
```

```python
X2.shape
```

```python
y = label
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain1,xtest1,ytrain,ytest = train_test_split(X1,y,test_size = 0.25,random_state = 1)
```

```python
xtrain2,xtest2,ytrain,ytest = train_test_split(X2,y,test_size = 0.25,random_state = 1)
```

```python
bnb = BernoulliNB()
```

```python
mnb = MultinomialNB()
```

```python
bnb.fit(xtrain1,ytrain)
```

```python
mnb.fit(xtrain2,ytrain)
```

```python
y_pred1 = bnb.predict(xtest1)
```

```python
y_pred2 = mnb.predict(xtest2)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(ytest,y_pred1)
```

```python
accuracy_score(ytest,y_pred2)
```
